```python
data1 = pd.read_excel('Datos/Alcaceres.xlsx')
data1
```

```python
# Empezar a limpiar
data1 = data1.fillna(' ')
data1.keys()
```

```python
# Ponemos el nombre completo de la persona en una sola columna
NombreCompleto = data1['Nombre 1']+' '+data1['Nombre 2']+' '+data1['Apellido 1']+' '+data1['Apellido 2']
data1.insert(loc=10, column='NombreCompleto',value=NombreCompleto)
```

```python
data1['NombreCompleto']
```

```python
data1 = data1.rename( columns={'Localidad (NOMBRE)':'Localidad'} )
data1
```

```python
Features = ['Tipo de Documento',\
            'Número de documento',\
            'NombreCompleto', \
            'Número de contacto 1',\
            'A que EPS se encuentra afiliado',\
            'Localidad',\
            'Fecha sesión ',\
            'Personas sesión ']
```

```python
for i in range(1,9):
  print(f'Processing session {i}')
  Columns = Features.copy()
  Columns[-1] += str(i)
  Columns[-2] += str(i)

  basemask = data1[Columns]

  mask = basemask[Columns[-1]] == 1
  # Filtro importante # Asiste la persona?

  basemask = basemask[mask]

  # Guardar la informacion
  print(f'Saving file for session {i}')
  output = pd.ExcelWriter('Alcaceres_Sesion_'+str(i)+'.xlsx', engine='xlsxwriter')
  basemask.to_excel(output, index=False, sheet_name='Hoja1')
  output.close()
```

```python
output = pd.ExcelWriter('Alcaceres.xlsx', engine='xlsxwriter')

for i in range(1,9):
  print(f'Processing session {i}')
  Columns = Features.copy()
  Columns[-1] += str(i)
  Columns[-2] += str(i)

  basemask = data1[Columns]

  mask = basemask[Columns[-1]] == 1
  # Filtro importante # Asiste la persona?

  basemask = basemask[mask]

  # Guardar la informacion
  print(f'Saving info for sheet {i}')
  basemask.to_excel(output, index=False, sheet_name='Hoja'+str(i))

output.close()
```
